<h1 align="center", style="color:green">Bar Chart</h1>

<h3 style = "color:purple">Simple bar chart showing revenues of major US tech companies</h3>

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
plt.bar(company,revenue)
plt.show()
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue, label="Revenue")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
plt.show()
```

<h3 style='color:purple'>Multiple Bars showing revenue and profit of major US tech companies</h3>

```python
profit=[40,2,34,12]
```

```python
plt.bar(xpos,revenue, width=0.4, label="Revenue")
plt.bar(xpos,profit, width=0.4,label="Profit")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
plt.show()
```

```python
plt.bar(xpos-0.2,revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2,profit, width=0.4,label="Profit")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
plt.show()
```

<h3 style = 'color:purple' >Horizontal bar chart using barh function</h3>

```python
plt.barh(xpos,revenue, label="Revenue")

plt.yticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technolog Stocks')
plt.legend()
plt.show()
```

<h1 align="center", style="color:green">Histogram</h1>

<h3 style = 'color:purple'>In histograms, x axis contains a variable and y axis will be a frequency of that variable</h3>

<h4 style = 'color:black'>We have a sample data of blood sugar level of different patients, we will try to plot number of patients by blood range and try to figure out how many patients are normal, pre-diabetic and diabetic</h4>

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, rwidth=0.5) # by default number of bins is set to 10
```

<h3 style='color:purple'>Bins Parameter</h3>

```python
plt.hist(blood_sugar,rwidth=0.5,bins=4)
```

<h3 style='color:purple'>Histogram showing normal, pre-diabetic and diabetic patients distribution</h3>

<ul style="color:brown;font-weight: bold;">
    <li>80-100: Normal</li>
    <li>100-125: Pre-diabetic</li>
    <li>80-100: Diabetic</li>
</ul>

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.8, color='g')
```

<h3 style='color:purple'>Mutiple data samples in a histogram</h3>

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150],
         rwidth=0.95, color=['green','orange'],label=['men','women'])
plt.legend()
```

<h3 style='color:purple'>histtype=step</h3>

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,histtype='step')
```

<h3 style='color:purple'>horizontal orientation</h3>

```python
plt.xlabel("Number Of Patients")
plt.ylabel("Sugar Level")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, orientation='horizontal')
```

<h1 align="center", style="color:green">Pie Charts</h1>

<h3 style="color:purple">Draw pie chart to track down home expenses</h3>

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

<h3 style="color:purple">Draw pie chart as perfect circle</h3>

```python
plt.pie(exp_vals,labels=exp_labels, shadow=True)
plt.axis("equal")
plt.show()
```

<h3 style="color:purple">Show percentages for every pie. Specify radius to increase chart size</h3>

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5)
plt.show()
```

<h3 style="color:purple">Explode</h3>

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2])
plt.show()
```

<h3 style="color:purple">counterclock and angle properties</h3>

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,
        explode=[0,0,0,0.1,0.2],counterclock=True, startangle=45)
plt.show()
```

<h2 style="color:green" align="center">Saving chart to a file using savefig</h2>

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%0.1f%%', explode=[0,0.1,0.1,0,0])
plt.savefig("piechart.jpg", bbox_inches="tight", pad_inches=1, transparent=True)
```

<h2 style= "color:purple">You can also save in pdf format and to a different location on your computer by specifying absolute path</h2>

```python
plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%0.1f%%', explode=[0,0.1,0.1,0,0])
plt.savefig("piechart.pdf", bbox_inches="tight", pad_inches=10, transparent=True)
```
